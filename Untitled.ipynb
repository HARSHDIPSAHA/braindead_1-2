```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'd' is your deliveries DataFrame

# Step 1: Filter deliveries for Powerplay (overs 1-6) and Death Overs (overs 16-20)
powerplay = d[(d['over'] >= 1) & (d['over'] <= 6)]
death_overs = d[(d['over'] >= 16) & (d['over'] <= 20)]

# Step 2: Calculate total runs in Powerplay and Death Overs for each match
powerplay_runs = powerplay.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index()
death_overs_runs = death_overs.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index()

# Step 3: Calculate average Powerplay and Death Overs scores for each team
avg_powerplay = powerplay_runs.groupby('batting_team')['total_runs'].mean().reset_index()
avg_death_overs = death_overs_runs.groupby('batting_team')['total_runs'].mean().reset_index()

# Rename columns for clarity
avg_powerplay.rename(columns={'total_runs': 'avg_powerplay_score'}, inplace=True)
avg_death_overs.rename(columns={'total_runs': 'avg_death_overs_score'}, inplace=True)

# Step 4: Merge the average scores into a single DataFrame
avg_scores = pd.merge(
    avg_powerplay,
    avg_death_overs,
    on='batting_team',
    how='outer'
).fillna(0)  # Fill NaN with 0 for teams with no data

# Step 5: Display the results
print(avg_scores)

# Step 6: Plot the average Powerplay and Death Overs scores with values on bars
plt.figure(figsize=(14, 6))

# Plot average Powerplay scores
plt.subplot(1, 2, 1)
ax1 = sns.barplot(x='batting_team', y='avg_powerplay_score', data=avg_scores, palette='viridis')
plt.title('Average Powerplay Score (Overs 1-6)', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Average Powerplay Score', fontsize=14)
plt.xticks(rotation=45)

# Add values on top of the bars for Powerplay
for p in ax1.patches:
    ax1.annotate(
        f'{p.get_height():.1f}',  # Display the value with 1 decimal place
        (p.get_x() + p.get_width() / 2., p.get_height()),  # Position of the text
        ha='center',  # Horizontal alignment
        va='center',  # Vertical alignment
        xytext=(0, 10),  # Offset for the text
        textcoords='offset points'
    )

# Plot average Death Overs scores
plt.subplot(1, 2, 2)
ax2 = sns.barplot(x='batting_team', y='avg_death_overs_score', data=avg_scores, palette='magma')
plt.title('Average Death Overs Score (Overs 16-20)', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Average Death Overs Score', fontsize=14)
plt.xticks(rotation=45)

# Add values on top of the bars for Death Overs
for p in ax2.patches:
    ax2.annotate(
        f'{p.get_height():.1f}',  # Display the value with 1 decimal place
        (p.get_x() + p.get_width() / 2., p.get_height()),  # Position of the text
        ha='center',  # Horizontal alignment
        va='center',  # Vertical alignment
        xytext=(0, 10),  # Offset for the text
        textcoords='offset points'
    )

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'd' is your deliveries DataFrame and 'm' is your matches DataFrame

# Step 1: Filter deliveries for Powerplay (overs 1-6)
powerplay = d[(d['over'] >= 1) & (d['over'] <= 6)]

# Step 2: Calculate total runs, wickets, boundaries, and dot balls in Powerplay for each match
powerplay_stats = powerplay.groupby(['match_id', 'batting_team']).agg(
    total_runs=('total_runs', 'sum'),
    wickets_lost=('is_wicket', 'sum'),
    boundaries=('batsman_runs', lambda x: x.isin([4, 6]).sum()),
    dot_balls=('total_runs', lambda x: (x == 0).sum())
).reset_index()

# Step 3: Calculate average Powerplay runs, wickets, boundaries, and dot balls for each team
avg_powerplay_stats = powerplay_stats.groupby('batting_team').agg(
    avg_runs=('total_runs', 'mean'),
    avg_wickets=('wickets_lost', 'mean'),
    avg_boundaries=('boundaries', 'mean'),
    avg_dot_balls=('dot_balls', 'mean')
).reset_index()

# Step 4: Calculate boundary and dot ball percentages
avg_powerplay_stats['boundary_percentage'] = (avg_powerplay_stats['avg_boundaries'] / 36) * 100  # 36 balls in Powerplay
avg_powerplay_stats['dot_ball_percentage'] = (avg_powerplay_stats['avg_dot_balls'] / 36) * 100

# Step 5: Display the results
print("Average Powerplay Stats:")
print(avg_powerplay_stats)

# Step 6: Plot average Powerplay runs
plt.figure(figsize=(12, 6))
sns.barplot(x='batting_team', y='avg_runs', data=avg_powerplay_stats, palette='viridis')
plt.title('Average Powerplay Runs by Team', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Average Powerplay Runs', fontsize=14)
plt.xticks(rotation=45)

# Add values on top of the bars
for p in plt.gca().patches:
    plt.gca().annotate(
        f'{p.get_height():.1f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='center',
        xytext=(0, 10),
        textcoords='offset points'
    )

plt.show()

# Step 7: Plot average Powerplay wickets lost
plt.figure(figsize=(12, 6))
sns.barplot(x='batting_team', y='avg_wickets', data=avg_powerplay_stats, palette='magma')
plt.title('Average Powerplay Wickets Lost by Team', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Average Powerplay Wickets Lost', fontsize=14)
plt.xticks(rotation=45)

# Add values on top of the bars
for p in plt.gca().patches:
    plt.gca().annotate(
        f'{p.get_height():.1f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='center',
        xytext=(0, 10),
        textcoords='offset points'
    )

plt.show()

# Step 8: Plot boundary and dot ball percentages
plt.figure(figsize=(14, 6))

# Boundary Percentage
plt.subplot(1, 2, 1)
sns.barplot(x='batting_team', y='boundary_percentage', data=avg_powerplay_stats, palette='coolwarm')
plt.title('Boundary Percentage in Powerplay', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Boundary Percentage', fontsize=14)
plt.xticks(rotation=45)

# Add values on top of the bars
for p in plt.gca().patches:
    plt.gca().annotate(
        f'{p.get_height():.1f}%',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='center',
        xytext=(0, 10),
        textcoords='offset points'
    )

# Dot Ball Percentage
plt.subplot(1, 2, 2)
sns.barplot(x='batting_team', y='dot_ball_percentage', data=avg_powerplay_stats, palette='coolwarm')
plt.title('Dot Ball Percentage in Powerplay', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Dot Ball Percentage', fontsize=14)
plt.xticks(rotation=45)

# Add values on top of the bars
for p in plt.gca().patches:
    plt.gca().annotate(
        f'{p.get_height():.1f}%',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='center',
        xytext=(0, 10),
        textcoords='offset points'
    )

plt.tight_layout()
plt.show()
```

```python
# Step 9: Top Performers in Powerplay
# Top Batsmen (most runs in Powerplay)
top_batsmen = powerplay.groupby('batter').agg(
    total_runs=('batsman_runs', 'sum'),
    balls_faced=('ball', 'count'),
    boundaries=('batsman_runs', lambda x: x.isin([4, 6]).sum())
).reset_index()
top_batsmen['strike_rate'] = (top_batsmen['total_runs'] / top_batsmen['balls_faced']) * 100
top_batsmen = top_batsmen.sort_values(by='total_runs', ascending=False).head(10)

print("Top Batsmen in Powerplay:")
top_batsmen
```

```python
# Top Bowlers (most wickets in Powerplay)
top_bowlers = powerplay[powerplay['is_wicket'] == 1].groupby('bowler').agg(
    wickets=('is_wicket', 'sum'),
    runs_conceded=('total_runs', 'sum'),
    balls_bowled=('ball', 'count')
).reset_index()
top_bowlers['economy'] = (top_bowlers['runs_conceded'] / top_bowlers['balls_bowled']) * 6
top_bowlers = top_bowlers.sort_values(by='wickets', ascending=False).head(10)

print("Top Bowlers in Powerplay:")
top_bowlers
```

```python
# Step 10: Powerplay Performance by Venue
powerplay_with_venue = pd.merge(
    powerplay_stats,
    m[['id', 'venue']],
    left_on='match_id',
    right_on='id'
)

venue_stats = powerplay_with_venue.groupby('venue').agg(
    avg_runs=('total_runs', 'mean'),
    avg_wickets=('wickets_lost', 'mean')
).reset_index()

print("Powerplay Performance by Venue:")
venue_stats
```

```python
# Step 11: Powerplay Performance by Phase (Overs 1-3 vs. Overs 4-6)
powerplay_phase_1 = powerplay[powerplay['over'] <= 3]
powerplay_phase_2 = powerplay[powerplay['over'] > 3]

phase_1_stats = powerplay_phase_1.groupby('batting_team').agg(
    avg_runs=('total_runs', 'mean'),
    avg_wickets=('is_wicket', 'mean')
).reset_index()

phase_2_stats = powerplay_phase_2.groupby('batting_team').agg(
    avg_runs=('total_runs', 'mean'),
    avg_wickets=('is_wicket', 'mean')
).reset_index()

print("Powerplay Performance by Phase (Overs 1-3):")
print(phase_1_stats)

print("Powerplay Performance by Phase (Overs 4-6):")
phase_2_stats
```

```python
# Step 12: Impact of Powerplay on Match Outcome
powerplay_with_result = pd.merge(
    powerplay_stats,
    m[['id', 'winner']],
    left_on='match_id',
    right_on='id'
)

powerplay_with_result['result'] = powerplay_with_result.apply(
    lambda row: 'Win' if row['batting_team'] == row['winner'] else 'Loss',
    axis=1
)

result_stats = powerplay_with_result.groupby(['batting_team', 'result']).agg(
    avg_runs=('total_runs', 'mean'),
    avg_wickets=('wickets_lost', 'mean')
).reset_index()

print("Impact of Powerplay on Match Outcome:")
result_stats
```

```python
# Assuming 'm' is your matches DataFrame and 'd' is your deliveries DataFrame

# Step 1: Merge 'm' and 'd' to get season information
merged_data = pd.merge(
    d,
    m[['id', 'season']],
    left_on='match_id',
    right_on='id'
)

# Step 2: Calculate total runs per match
runs_per_match = merged_data.groupby(['season', 'match_id'])['total_runs'].sum().reset_index()

# Step 3: Calculate average runs per match per season
avg_runs_per_match = runs_per_match.groupby('season')['total_runs'].mean().reset_index()

# Step 4: Display the results
print("Average Runs per Match per Season:")
avg_runs_per_match
```

```python
# Step 1: Filter matches where the target is 200+ runs
high_target_matches = m[m['target_runs'] >= 200]

# Step 2: Group by season and count the number of 200+ targets
high_targets_per_season = high_target_matches.groupby('season').size().reset_index(name='200+_targets')

# Step 3: Display the results
print("Targets of 200+ Runs per Season:")
high_targets_per_season
```

```python
# Step 1: Calculate total runs scored by each team per season
team_runs_per_season = merged_data.groupby(['season', 'batting_team'])['total_runs'].sum().reset_index()

# Step 2: Calculate the number of matches played by each team per season
matches_per_team_per_season = merged_data.groupby(['season', 'batting_team'])['match_id'].nunique().reset_index()

# Step 3: Calculate average runs per team per season
avg_runs_per_team_per_season = pd.merge(
    team_runs_per_season,
    matches_per_team_per_season,
    on=['season', 'batting_team']
)
avg_runs_per_team_per_season['avg_runs'] = avg_runs_per_team_per_season['total_runs'] / avg_runs_per_team_per_season['match_id']

# Step 4: Display the results
print("Average Score of Each Team per Season:")
avg_runs_per_team_per_season[['season', 'batting_team', 'avg_runs']]
```

```python
# Step 1: Calculate total runs scored by each batsman per season
batsman_runs_per_season = merged_data.groupby(['season', 'batter'])['batsman_runs'].sum().reset_index()

# Step 2: Identify the Orange Cap holder (top run-scorer) for each season
orange_cap_holders = batsman_runs_per_season.loc[
    batsman_runs_per_season.groupby('season')['batsman_runs'].idxmax()
]

# Step 3: Display the results
print("Runs of Orange Cap Holders per Season:")
orange_cap_holders
```

```python
# Step 1: Calculate total wickets taken by each bowler per season
bowler_wickets_per_season = merged_data[merged_data['is_wicket'] == 1].groupby(['season', 'bowler']).size().reset_index(name='wickets')

# Step 2: Identify the Purple Cap holder (top wicket-taker) for each season
purple_cap_holders = bowler_wickets_per_season.loc[
    bowler_wickets_per_season.groupby('season')['wickets'].idxmax()
]

# Step 3: Display the results
print("Wickets of Purple Cap Holders per Season:")
purple_cap_holders
```

```python
# Step 1: Calculate total wickets and economy rate for each bowler per season
bowler_stats_per_season = merged_data.groupby(['season', 'bowler']).agg(
    wickets=('is_wicket', 'sum'),
    runs_conceded=('total_runs', 'sum'),
    balls_bowled=('ball', 'count')
).reset_index()
bowler_stats_per_season['economy'] = (bowler_stats_per_season['runs_conceded'] / bowler_stats_per_season['balls_bowled']) * 6

# Step 2: Get the top 10 bowlers per season based on wickets
top_10_bowlers_per_season = bowler_stats_per_season.sort_values(
    by=['season', 'wickets', 'economy'],
    ascending=[True, False, True]
).groupby('season').head(10)

# Step 3: Display the results
print("Top 10 Bowlers per Season:")
top_10_bowlers_per_season[['season', 'bowler', 'wickets', 'economy']]
```

```python
# Step 1: Calculate batting average and strike rate for the top 20 run-scorers
# Batting Average = Total Runs / Number of Dismissals
# Strike Rate = (Total Runs / Total Balls Faced) * 100

# Calculate total runs, balls faced, and dismissals for each batsman
batsman_stats = d.groupby('batter').agg(
    total_runs=('batsman_runs', 'sum'),
    balls_faced=('ball', 'count'),
    dismissals=('is_wicket', 'sum')
).reset_index()

# Filter for the top 20 run-scorers
batsman_stats_top_20 = pd.merge(
    top_20_run_scorers,
    batsman_stats,
    on='batter'
)

# Calculate batting average and strike rate
batsman_stats_top_20['batting_average'] = batsman_stats_top_20['total_runs'] / batsman_stats_top_20['dismissals']
batsman_stats_top_20['strike_rate'] = (batsman_stats_top_20['total_runs'] / batsman_stats_top_20['balls_faced']) * 100

# Step 2: Plot Batting Average vs Strike Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='batting_average', y='strike_rate', data=batsman_stats_top_20, hue='batter', s=100)
plt.title('Batting Average vs Strike Rate for Top 20 Run-Scorers', fontsize=16)
plt.xlabel('Batting Average', fontsize=14)
plt.ylabel('Strike Rate', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Calculate bowling economy rate
bowling_stats = d.groupby('bowler').agg(
    total_runs_conceded=('total_runs', 'sum'),
    total_overs=('over', 'nunique')
).reset_index()

# Handle division by zero for bowling economy
bowling_stats['bowling_economy'] = np.where(
    bowling_stats['total_overs'] == 0,  # Check if total_overs is zero
    0,  # Set economy to 0 if total_overs is zero
    bowling_stats['total_runs_conceded'] / bowling_stats['total_overs']  # Otherwise, calculate economy
)

# Step 2: Merge batting and bowling stats
# Ensure 'player_stats' is already calculated (from previous steps)
all_rounder_stats = player_stats.merge(bowling_stats, left_on='batter', right_on='bowler', how='inner')

# Step 3: Handle infinite values in batting_average and bowling_economy
# Replace infinite values with NaN
all_rounder_stats.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with NaN values in 'batting_average' or 'bowling_economy'
all_rounder_stats = all_rounder_stats.dropna(subset=['batting_average', 'bowling_economy'])

# Step 4: Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
all_rounder_stats['cluster'] = kmeans.fit_predict(all_rounder_stats[['batting_average', 'bowling_economy']])

# Step 5: Plot the results
plt.figure(figsize=(10, 6))
sns.scatterplot(data=all_rounder_stats, x='bowling_economy', y='batting_average', hue='cluster', palette='Set1', s=100)
plt.title('Batting Average vs Bowling Economy Rate (Clustered)')
plt.xlabel('Bowling Economy Rate')
plt.ylabel('Batting Average')
plt.legend(title='Cluster')
plt.show()
```

```python
# Create the required dataframe
df = m[['team1', 'team2', 'venue', 'toss_winner', 'toss_decision', 'winner']]

# Rename columns for clarity
df.columns = ['team1', 'team2', 'venue', 'won_toss', 'choose_to', 'winner']

# Add match_type column directly from m
if 'match_type' in m.columns:
    df['match_type'] = m['match_type']
else:
    df['match_type'] = 'league'  # Default to league matches if column is missing

# Add target column
# Target = 1 if team1 is the winner, 0 otherwise
df['target'] = (df['team1'] == df['winner']).astype(int)
```

```python
# Calculate head-to-head win rates
head_to_head = {}

# Iterate through all matches to calculate head-to-head stats
for index, row in m.iterrows():
    team_a = row['team1']
    team_b = row['team2']
    winner = row['winner']

    # Create a unique key for the team pair (sorted to avoid duplicates like MI-CSK and CSK-MI)
    key = tuple(sorted([team_a, team_b]))

    # Initialize head-to-head stats if not already present
    if key not in head_to_head:
        head_to_head[key] = {'total_matches': 0, team_a: 0, team_b: 0}

    # Update total matches and wins for each team
    head_to_head[key]['total_matches'] += 1
    if winner == team_a:
        head_to_head[key][team_a] += 1
    elif winner == team_b:
        head_to_head[key][team_b] += 1

# Add head-to-head win rates to the dataframe
df['team1_head_to_head'] = 0.0
df['team2_head_to_head'] = 0.0

for index, row in df.iterrows():
    team_a = row['team1']
    team_b = row['team2']
    key = tuple(sorted([team_a, team_b]))

    if key in head_to_head:
        total_matches = head_to_head[key]['total_matches']
        if total_matches > 0:
            df.at[index, 'team1_head_to_head'] = head_to_head[key][team_a] / total_matches
            df.at[index, 'team2_head_to_head'] = head_to_head[key][team_b] / total_matches
```
